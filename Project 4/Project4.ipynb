## Join Queries

7. __Perform__ the following __SQL Join Queries__ :

__Join "Movies" and "Votes"__ (all columns).

```python
con = sq3.connect('movies.db')
```

```python
pd.read_sql('SELECT * FROM Movies m left join Votes v on m.id=v.id', con)
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average").__

```python
pd.read_sql('SELECT m.id, m.title, v.vote_average FROM Movies m \
              LEFT JOIN Votes V \
              ON m.id=v.id', con)
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average") and return only those movies with vote_average > 8.__

```python
pd.read_sql('SELECT m.id, m.title, v.vote_average \
             FROM Movies m \
             LEFT JOIN Votes v \
             ON m.id=v.id \
             WHERE v.vote_average > 8',
            con, index_col='id'
           )
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average") and return only those movies with vote_average > 8 and in ascending budget order__.

```python
pd.read_sql('SELECT m.id, m.title, v.vote_average \
             FROM Movies m \
             LEFT JOIN Votes v \
             ON m.id=v.id \
             WHERE v.vote_average > 8 \
             ORDER BY m.Budget',
            con, index_col='id'
           )
```

8. __Perform__ the following advanced __SQL Queries__ and __store__ the results in DataFrames:

__Get the Total Revenue (sum) for each Production Company.__

```python
df = pd.read_sql('SELECT p.id, p.comp_name, m.revenue, m.title \
            FROM Prod p \
            LEFT JOIN Movies m \
            ON m.id = p.id', con)

df.groupby('comp_name').revenue.sum().sort_values(ascending=True)
```

```python
pd.read_sql('SELECT sum(revenue), p.comp_name \
            FROM Prod p \
            LEFT JOIN Movies m \
            ON m.id = p.id \
            GROUP BY p.comp_name \
            ORDER BY 1 ', con)
```

__Get all Production Companies for the movie "Titanic".__

```python
pd.read_sql('SELECT m.title, p.comp_name \
             FROM Movies m \
             LEFT JOIN Prod p \
             ON m.id=p.id \
             WHERE m.title="Titanic"', con)
```

__Get the Total Revenue (sum) for each Genre.__

```python
pd.read_sql('SELECT sum(m.revenue), g.genres_name \
            FROM Movies m \
            LEFT JOIN genres g \
            GROUP BY g.genres_name', con)
```
